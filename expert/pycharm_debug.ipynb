# PyCharm调试器使用说明

PyCharm提供了图形化的、完善的调试功能，支持断点设置，单步模式，表达式求值，变量查看等功能，特别是支持多线程，远程调试等pdb不具备的功能，

在运行调试之前，复习一下在PyCharm中设置运行/调试配置文件的相关知识。也就是说，在运行或调试程序之前，先设置运行/调试配置文件。

仍然使用pdb用过的程序来演示使用PyCharm调试器的使用：

```python
%%writefile debug_demo01.py
#!/usr/bin/env python
# -*- coding: utf-8 -*-
# Copyright (c) 2018--2023, Wang Weihua.
# All rights reserved.
"""pdb demo program"""


class SimpleClass(object):
    """simple class"""
    def __init__(self):
        self.number = 0

    def increment(self):
        """increment 1"""
        self.number += 1


def afunc(var):
    """a function"""
    result = 0
    print(type(var), id(var), var)
    if isinstance(var, int):
        for i in range(var):
            result += i

    return result


def main():
    """main entry"""
    result = afunc(5)
    sa = SimpleClass()
    sa.increment()
    print(result, sa.number)


if __name__ == '__main__':
    main()
```

## 设置断点

在PyCharm中调试程序，首先要设置断点。选中对应代码行，有两种方法可以设置断点：
- 单击左边框的空白处
- 按下`Ctrl+F8`快捷键

例如在第20行设置断点，结果如下图所示：
![设置断点](assets/pycharm_debug_break.png)

选中已设置断点的语句行，再次单击左边框的空白处或者按`Ctrl+F8`快捷键，会取消断点设置

在第24行语句也设置断点，然后按`Ctrl+Shift+F8`快捷键，会弹出窗口用来断点修改窗口，在该窗口可以设置条件断点，如`result == 3`。

![条件断点](assets/pycharm_debug_break_condtion.png)

可以看出断点图标已经有所改变，如下图所示：
![断点图标](assets/pycharm_debug_breaks.png)

使用快捷键`Ctrl+Shift+F8`还可以用来查看所有断点。
![查看全部断点](assets/pycharm_debug_breakpoints.png)

## 调试运行

设置好断点后，单击工具栏中的绿色蜘蛛按钮，或者按`Shift+F9`开始调试，并在第一个断点出停止。

![运行调试](assets/pycharm_debug_start.png)

在控制台窗口有下图如下：
![调试控制台](assets/pycharm_debug_windows.png)

控制台窗口有如下内容：
- Debugger页
    - Frames
    - Variables
- Console页
    - ipython
- 工具栏
- Debug控制按钮

主要运行快捷键：
- `F8`，步进（step over）
- `F7`，进入（step into）函数体
- `Shift+F8`，跳出（step out）函数体
- `Alt+F9`,运行到当前光标
