# PyCharm使用说明

PyCharm是专门用于Python开发的集成开发环境，对Django、Flask等Web开发有特别支持。除此之外，PyCharm缺省还支持HTML5、CSS、JavaSscript等语言。通过安装其他插件还可以支持其他语言。本节是Pycharm的使用说明。

PyCharm是一个专业的集成开发环境，支持如下功能：
- 项目管理
- 语法高亮
- 代码跳转
- 智能提示
- 自动完成
- 调试
- 单元测试
- 版本控制

## 安装

Pycharm是个跨平台开发环境，可以运行在Windows、Linux、Mac等常用操作系统。

Pycharm包括社区版（Community）和专业版（Professional）。前者免费，后者需要收费。

### 下载安装包

浏览[PyCharm官网](https://www.jetbrains.com/pycharm/)，找到下载按钮并单击，切换到[下载页面](https://www.jetbrains.com/pycharm/download/)。

根据自身操作系统，选择下载版本，社区版或专业版。这里选择社区版。

### Windows系统

切换到下载目录，双击下载安装程序`pycharm-community-2018.2.1.exe`，开始安装。

使用缺省值，一路安装下去。Pycharm程序安装目录可以进行更改：
![Pycharm安装目录](assets/pycharm_windows_install.png)

安装完成后，Pycharm第一次启动时，也使用缺省选择，例如；
![Pycharm首次启动选择](assets/pycharm_windows_conplete_installation.png)

最后进入欢迎界面
![首次使用](assets/pycharm_welcome.png)

### Linux系统

切换到下载目录，解压缩文件：
```sh
tar zxvf pycharm-professional-2018.2.tar.gz
```

切换到解压缩后文件，再切换到bin目录,执行命令：
```sh
bash pycharm.sh
```

为了方便使用在用户初始化文件`.bashrc`中，添加
```
alias pycharm='/usr/local/pycharm-2018.2/bin/pycharm.sh'
```

## 快速开始

下面介绍如何快速使用PyCharm软件开发Python程序。示例使用平台为Windows系统。

### 打开或创建项目

在PyCharm环境中，所有事情都是在项目（Project）的上下文中去完成的。也就是说，项目是智能助手、重构、编码风格兼容等的一个基础。

在PyCharm启动后的欢迎页面中，提示有三种方法来创建或打开一个项目：
- 打开已有项目
- 检出已有项目
- 创建新项目

![欢迎页面](assets/pycharm_welcome.png)

#### 创建新项目

在欢迎界面选择【Create New Project】从零开始创建新项目。在弹出页面，首先选择项目位置，这里指定为`D:\pywork\pycharm-first`。

选择项目使用的Python解释器，有两种选项：
- 指定已有Pyhton解释器
- 新建Python虚拟环境
    - Virtualenv
    - Pipenv
    - Conda

这里选择使用Pipenv来创建Python虚拟环境。
![创建新项目](assets/pycharm_create_project_pipenv.png)

PyCharm会会创建`pycharm-first`项目，同时使用Pipenv创建Python虚拟环境。项目创建完成后，出现如下界面
![项目创建完成](assets/pycharm_project_created.png)

#### 打开已有项目

在欢迎页面选择【Open】，然后从用户目录中选择项目全路径。例如这里选择为`D:\pywork\pycharm-hello`，然后单击【OK】确认。

![打开已有项目](assets/pycharm_opening_project.png)

同样PyCharm会为源码目录创建项目。

#### 检出已有项目

在欢迎页面选择【Check out from Version Control】下来菜单，从中显示【Git】，然后从用户目录中选择项目全路径。例如这里选择为`D:\pywork\pycharm-hello`，然后单击【OK】确认。

PyCharm会克隆远程仓库，同时也会创建一个项目。

#### 定制项目的Python解释器

Python项目开发总是要有一个Python解释器。在PyCharm中可以为项目定制Python解释器。选择【File】=>【Settings】（`Ctrl+Alt+S`），在弹出配置窗口，选择【Project Interpreter
】会弹出Python解释器配置窗口。

![定制Python解释器](assets/pycharm_settings_project_intepreter.png)

在弹出窗选择【Project Interpreter】，单击项目解释器一栏最右边的按钮，并选择【Add...】，会弹出`Add Python Interpreter`窗口。

![添加Python解释器](assets/pycharm_settings_project_intepreter_add.png)

可以添加如下Python解释器：
- 系统解释器
- Pipenv虚拟环境
- Virtualenv虚拟环境
- Conda虚拟环境

在Pycharm专业版本中，还支持
- 远程Python解释器（SSH、Vagrant、WSL）
- 基于Docker的解释器（Docker、Docker Compose）

#### `.idea`目录

当PyCharm创建一个项目后，会在项目中创建一个`.idea`的文件夹，用于PyCharm的项目管理。

### PyCharm窗口浏览

打开一个项目后，会看到一个分为多个工作区域的主窗口，如下图所示。

![PyCharm工作区域](assets/pycharm_lookAroundThumb.png)

1. 项目视图，显示项目文件
2. 编辑器，代码编写的主战场。可以使用标签在打开文件中导航。
3. 导航栏，用于快速运行、调试、以及版本控制操作
4. 左边框，用于断点显示、代码导航、代码行、版本历史
5. 右边框，代码审查显示，如错误、经过等。
6. 工具窗口，绑定专业工具
7. 状态栏，指示项目与IDE状态，显示警告，以及文件编码、行分隔符、断面检查等信息。

在PyCharm窗口左下角，状态栏最左边，有一个按钮。鼠标在按钮上悬停一些时间，弹出一个列表窗口，列出了当前窗口有效工具列表来。

![PyCharm有效工具列表](assets/pycharm_ToolWindowQuickAccess.png)

### 定制环境

PyCharm高度可定制化，可以定制出符合自己使用习惯的开发环境。选择【File】=>【Settings】（`Ctrl+Alt+S`），会弹出配置窗口。下面介绍一些可定制内容选项。

#### 外观（Appearance）

在配置窗口，选择【Appearance & Behavior】=>【Appearance】，可以选择不同的IDE主题，缺省是`Darcula`主题。

![外观定制](assets/pycharm_settings_appearance.png)

#### 编辑器（Editor）

在配置窗口，选择【Editor】，有如下内容可供定制：
- 通用
- 字体
- 颜色
- 代码风格
- 代码审查（Inspections）

- 文件和代码模板（File and Code Templates）
- 文件编码（File Encodings）
- Live Templates
- 文件类型（File Types）
- Emmet

- 图像（Images）
- 意向（Intentions）
- Language Injections
- 拼写（Spelling）
- TODO

#### 快捷键

PyCharm是以快捷键为中心的使用方法，其中大多数操作都对应有快捷键。

在配置窗口，选择【Keymap】可以更改快捷键风格，也可以修改某些快捷键。这里使用缺省。

![快捷键定制](assets/pycharm_settings_keymap.png)

#### 其它

- 版本控制（Version Control）

- 编译、执行、部署（Build、Execution，Deployment）
- 语言与框架（Languages&Frameworks）
- 工具（Tools）

#### Intention actions

### 保护代码简洁

PyCharm会监视代码，并试图保持代码的正确与简洁。代码审查会检测存在的潜在错误与问题，并建议修改。

集成开发环境会发现无用代码、无限循环或其它需要注意的事情，会使用灯泡图标来提示，使用快捷键`Alt+Enter`会进行修复。

### 代码自动生成

使用PyCharm能自动生成一些结构化代码，这有助于加快编写速度，提高效率。在【Code】=>【Generate】（`Alt+Insert`）会帮助创建代码。

### 查找搜索

当项目大到一定程度，或者在阅读他人代码时，准确快速查找内容，对深入研究代码至关重要。

####  基本搜索

可以在指定范围内，查找替换任何片段代码，包括：
- 当前打开文件（`Ctrl+F`）
- 整个项目（`Ctrl+Shift+F）`）

### 运行、调试与测试

初始时导航栏上的【运行】和【调试】都是灰色的，是无法点击使用的。要想使用，需要先创建一个运行调试配置。在PyCharm中要想对代码进行运行、调试、测试操作时，实际上开启了基于配置的一个进程。

选择【Run】=>【Edit Configurations】会打开【Run/Debug Configurations】窗口。在窗口单击【加号】按钮添加一个配置。

在配置界面需要指定如下：
- 配置名字
- 脚本路径
- 指定脚本参数（Script parameters）
- Python解释器

其它还包括
- 环境变量
- 解释器选项
- 工作路径

![添加运行调试配置](assets/pycharm_configurations_add.png)

创建运行调试配置后，使用快捷键`Shift+F10`可以运行该配置。

#### 运行

在代码编辑区域按右键，选择**Run [name]**（或`Ctrl+Shift+F10`），会快捷创建一个配置文件，同时运行配置文件。

如果Python脚本中包含**__main__**语句，在左边框会有一个执行按钮，单击并选择相应命令。

#### 调试

如果程序在运行时出现错误异常，可以使用调试来查找问题解决问题。

在程序代码中，找到期望终端暂停的代码行，在其左边框处单击，就会出现断点标识。

使用快捷键`Shift+F9`，可以快速启动程序的调试。使用**Run**菜单下选项或调试工具窗口按钮，就可以单步执行、计算表达式，添加变量监视等。

![调试窗口](assets/pycharm_configurations_debug.png)

#### 测试

测试自己的程序总是不错，使用PyCharm会让测试变得简单。

PyCharm支持如下测试框架：
- Doctest
- Unittest
- Nosetest
- py.test

### 代码版本控制

使用版本控制系统来管理代码应该成为程序员的标配。PyCharm集成了常见的版本控制系统，如Git等。选择【File】=>【Settings】=>【Version Control】进行配置。

选择【VCS】【Show Changes】（`Alt + 9`）用来查看更动。

### 远程开发

在PyCharm中，可以使用远程解释器来运行项目。

需要添加和配置如下远程解释器：
- SSH
- Vagrant
- Docker
- Docker Composer

### 数据库

PyCharm自带了一个简单的数据库插件，可以比较方便的进行简单的数据库操作。
1. 创建，修改和删除数据表，字段，索引，主键，外键等。
2. 提供table editor来进行数据操作
3. 提供console来运行sql命令
4. 提供数据导出功能
