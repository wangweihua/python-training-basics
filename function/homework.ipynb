### 练习

使用递归函数实现一个数的整数幂。

```python
def power(x, n):
    """return x ** n"""
    result = 1
    for i in range(n):
        result *= x
        
    return result
```

```python
power(2, 3)
```

```python
def power2(x,n):
    if n == 0:
        return 1
    else:
        return x * power2(x, n-1)
```

```python
power2(2, 3)
```
